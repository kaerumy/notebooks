```python
import requests
import json

#Dewan Rakyat MP Posts in Sinar Malaysia Popit Database
posts = []
for page in range(1,10):
    dewan_rakyat_request = requests.get('http://sinar-malaysia.popit.mysociety.org/api/v0.1/search/posts?q=organization_id:53633b5a19ee29270d8a9ecf'+'&page='+str(page))
    for post in (json.loads(dewan_rakyat_request.content)['result']):
        posts.append(post)
        
import pandas
post_data = { "area_id": [], "id": [], "name": [] }
post_data_index = { "id": [], "area_id": [] }
for post in posts:
    post_data["id"].append(post["id"])
    post_data_index["id"].append(post["id"])
    post_data['area_id'].append(post["area"]['id'])
    post_data_index["area_id"].append(post["area"]['id'])
    post_data['name'].append(post['area']['name'])
 
posts_data = { "id": pandas.Series(post_data["id"], index=post_data_index["id"]),
                "area_id": pandas.Series(post_data["area_id"], index=post_data_index["area_id"]),
                "name": pandas.Series(post_data['name'])
             }
pandas.set_option('display.max_rows',1000)
posts_df = pandas.DataFrame(post_data)
posts_df.sort('area_id')
```

```python
#Current MPs

import datetime
from dateutil import parser

current = datetime.date(2013,5,5)
member_good = []
member_total = []

def member_info(count,post):
    
    if count > 1:
        info = post['area']['id'] + " " + post['area']['name'] + " " + post['id'] +  ": " + "Too many members"
    if count < 1:
        info = post['area']['id'] + " " + post['area']['name'] + " " + post['id'] +  ": " + "No members"
    if count == 1:
        info = None
    
    return info

def has_end_date(member):
    
    if member.has_key('end_date') and member['end_date'] == '':
        return False
    elif not member.has_key('end_date'):
        return False
    else:
        return True
    
def current_MP(member):
    #Legislative tag term here would simply
    if (parser.parse(member['start_date'])).date() > current:
        if not has_end_date(member):
            return True
    else:
        return False
    
    
for post in posts:
    for member in post['memberships']:
        if not member.has_key('start_date'):
            print member['id']
           
    
for post in posts:
    
    #current
    count = sum(1 for member in post['memberships'] if current_MP(member) )
    
    if member_info(count,post):
        print member_info(count,post)
    
    if count == 1:
        member_good.append(post['id'])
            
    member_total.append(count)
               
print len(member_good)
print sum(member_total)
```

```python
def person(person_id):
    #Load up information of persons from Popit database
    req = requests.get('https://sinar-malaysia.popit.mysociety.org/api/v0.1/persons/' + person_id)
    return json.loads(req.content)['result']

MP_ids = []

for post in posts:
    for member in post['memberships']:
        if current_MP(member):
            MP_ids.append(member['person_id'])

len(MP_ids)
            
```
