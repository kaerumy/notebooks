```python
import requests
import json

organizations = []

for id in df.index.values:
    rp = requests.get('http://sinar-malaysia.popit.mysociety.org/api/v0.1/persons/'+id)
    person = json.loads(rp.content)['result']
    memberships = person['memberships']
    
    if memberships:
        
        for member in memberships:
                
            ro_url = 'http://sinar-malaysia.popit.mysociety.org/api/v0.1/organizations/' + member['organization_id']
            ro = requests.get(ro_url)
            org = json.loads(ro.content)['result']
            
            try:
                del post
                
            except NameError:
                pass
            
            if member.has_key('post_id'):
                if member['post_id']:
                    rp_url = 'http://sinar-malaysia.popit.mysociety.org/api/v0.1/posts/' + member['post_id']
                    rp = requests.get(rp_url)
                    post = json.loads(rp.content)['result']
        
            if not member.has_key('role'):
                member['role'] = ''
            
            try:
                post
            except NameError:
                post = { 'label' : '' }
                
            print id + ", " + person['name'] + ', ' + org['name'] + ', ' + member['role'] + ', ' + post['label'] + '.'
        
            politicians = []    
            politicians.append({ 'popit_id' : id,  
                                   'name' : person['name'],
                                   'org'  : org['name'],
                                   'role' : member['role'], 
                                   'post': post['label'],
                                })
```
